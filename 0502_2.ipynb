```python
import csv
import matplotlib.pyplot as plt

f=open('훈영씨.csv', 'r', encoding='cp949')
data=csv.reader(f)
header=next(data)

avg_T=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[2] != '':
        date=row[0].split('-')
        avg_T[int(date[1])-1].append(float(row[2]))

f.close()

plt.title('')
plt.boxplot(avg_T)
```
