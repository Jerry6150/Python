```python
import csv
import matplotlib.pyplot as plt

f=open('훈영씨.csv', 'r', encoding='cp949')
data=csv.reader(f)
header=next(data)
print(header)

high_temp=[]
low_temp=[]

for row in data:
    if row[-1] != '' and row[-2] != '':
        date =row[0].split('-')
        if date[1] =='06' and date[2] =='04':
            high_temp.append(float(row[-1]))
            low_temp.append(float(row[-2]))

f.close()

plt.title('')
plt.plot(high_temp, 'yellow', label='high')
plt.plot(low_temp, 'green', label='low')
plt.legend(loc=0)
```
