```python
import csv
import matplotlib.pyplot as plt

f=open('.csv', 'r', encoding='cp949')
data=csv.reader(f)
header=next(data)

avg_temp=[]
for i in range(31):
    avg_temp.append([])

for row in data:
    if row[2] != '':
        date = row[0].split('-')
        month = date[1] #  
        day = date[2] #  
        if month == '08':
            x = int(day)-1
            avg_temp[x].append(float(row[2]))

f.close()

plt.style.use('ggplot')
plt.figure(figsize=(16,9), dpi=300)
plt.title('Daejeon daily average temperature distribution')
plt.boxplot(avg_temp, showfliers=False)
```
