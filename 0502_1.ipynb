```python
import csv
import matplotlib.pyplot as plt

f=open('훈영씨.csv', 'r', encoding='cp949')
data=csv.reader(f)
header=next(data)

jan=[]
aug=[]

for row in data:
    if row[-1] != '':
        date=row[0].split('-')
        if date[1]=='01':
            jan.append(float(row[-1]))
        if date[1]=='08':
            aug.append(float(row[-1]))

f.close()

plt.title('')
plt.hist(jan, color='yellow', bins=100, label='jan')
plt.hist(aug, color='green', bins=100, label='aug')
plt.legend()

```
